* Create a Tree like Structure for an organization and print it in a pretty way

```scala
import scala.collection.mutable.ListBuffer
import scala.language.implicitConversions
case class Employee(id:Int,name:String,department:String,salary:Double)
case class Result(success:Boolean=true,message:String="success")


implicit def tupleToEmployee(tup:(Int,String,String,Double)):Employee=Employee(tup._1,tup._2,tup._3,tup._4)

//the same department cannot be at two different levels for the same tree

class Tree(val department:String="ORGANIZATION"){


    val departments:ListBuffer[Tree]=ListBuffer.empty[Tree]
    val employees:ListBuffer[Employee]=ListBuffer.empty[Employee]


    def print():Unit={
        
        println("Printing the organization structure")
    }
    def addRecord(parentOrg:Option[String], org:Option[String], employee:Option[Employee]):Result={
            
            
           //probably dfs first for parent Org
           //checking for the cases when there's nothing to be added
           if(parentOrg.isEmpty && org.isEmpty){
               Result()//nothing to add
           }
           else if(parentOrg.isDefined && org.isEmpty){
             Result(false,"Department name cannot be empty")
           }
           else if(parentOrg.isEmpty && !org.isEmpty ){

                if(!org.get.equalsIgnoreCase("ORGANIZATION")){

                        this.addEmployee(this,employee)

                }
                Result(false,"Cannot add department to the root")
               
           }
           else{

                val parent:Option[Tree]=this.findDepartment(parentOrg.get)
                println(s"Parent is ${parent.getOrElse(new Tree("Not found")).department}") 
                if(parent.isDefined){
                        
                        val curOrg:Option[Tree]=this.addOrGetDepartment(parent.get,org.get)
                        this.addEmployee(curOrg.get,employee)
                        Result()

                }
                else{
                    Result(false,"Parent department not found")
                }
           }
           


    }
    def addEmployee( org:Tree, employee:Option[Employee]):Boolean={

        if(employee.isDefined){
            org.employees.append(employee.get)
            true
        }
        false

    }
    def addOrGetDepartment( parOrg:Tree, department:String):Option[Tree]={

        val dept:Option[Tree]=parOrg.departments.find(_.department.equalsIgnoreCase(department))
        if(dept.isEmpty){
            val deptFin:Tree=new Tree(department)
            parOrg.departments.append(new Tree(department))
            Some(deptFin)
        }
        else{
            dept
        }
    }

    def findDepartment(department:String):Option[Tree]={

        

        if(this.department.equalsIgnoreCase(department)){
            println(s"Found ${department}")
            Some(this)
        }
        else{
            for(i<-0 until this.departments.length){
                val temp:Option[Tree]=departments(i).findDepartment(department)
                if(temp.isDefined){
                     //->Returning if parent is found
                     println(s"Found dept, parent is ${this.department} ,org is ${temp.get.department}")
                    return temp
                }
            }
            if(this.department.equalsIgnoreCase("ORGANIZATION") && department.equalsIgnoreCase("HR")){
                println("lol")
            }
            
            None
        }
    }



}


```
